```python
import random
import csv
import os
import pandas as pd
import pickle
from datetime import datetime
from math import cos, asin, sqrt
from typing import List
import numpy as np
import scipy.cluster.vq as vq
from sklearn.cluster import KMeans
import cplex

import numpy as np
```

```python
class BaseStation:


    def __init__(self, id, addr, lat, lng):
        self.id = id
        self.address = addr
        self.latitude = lat
        self.longitude = lng
        self.num_users = 0
        self.workload = 0

    def __str__(self):
        return "No.{0}: {1}".format(self.id, self.address)
    
class EdgeServer(object):
    
    def __init__(self, id, latitude, longitude, base_station_id=None):
        self.id = id
        self.latitude = latitude
        self.longitude = longitude
        self.base_station_id = base_station_id
        self.assigned_base_stations = []
        self.workload = 0
```

```python
class DataUtils(object):
    def __init__(self, location_file, user_info_file):
        self.base_stations = self.base_station_reader(location_file)
        self.base_stations = self.user_info_reader(user_info_file)
        self.distances = self.distance_between_stations()


    def base_station_reader(self, path: str):
        bs_data = pd.read_csv(path, header=0, index_col=0)
        base_stations = []
        for index, bs_info in bs_data.iterrows():
            base_stations.append(BaseStation(id=index, addr=bs_info['address'], lat=bs_info['latitude'], lng=bs_info['longitude']))
        return base_stations


    def user_info_reader(self, path: str) -> List[BaseStation]:
     
        self.address_to_id = {bs.address: bs.id for bs in self.base_stations}

        req_data = pd.read_csv(path, header=0, index_col=0)
        req_data['start time'] = pd.to_datetime(req_data['start time'])
        req_data['end time'] = pd.to_datetime(req_data['end time'])
        for index, req_info in req_data.iterrows():
            service_time = (req_info['end time'] - req_info['start time']).seconds / 60
            bs_id = self.address_to_id[req_info['address']]
            self.base_stations[bs_id].num_users += 1
            self.base_stations[bs_id].workload += service_time
        return self.base_stations


    def _shuffle(l: List):
        random.seed(6767)
        random.shuffle(l)


    def calc_distance(lat_a, lng_a, lat_b, lng_b):

        p = 0.017453292519943295  # Pi/180
        a = 0.5 - cos((lat_b - lat_a) * p) / 2 + cos(lat_a * p) * cos(lat_b * p) * (1 - cos((lng_b - lng_a) * p)) / 2
        return 12742 * asin(sqrt(a))  # 2*R*asin...



    def distance_between_stations(self) -> List[List[float]]:

        assert self.base_stations
        base_stations = self.base_stations
        distances = []
        for i, station_a in enumerate(base_stations):
            distances.append([])
            for j, station_b in enumerate(base_stations):
                dist = DataUtils.calc_distance(station_a.latitude, station_a.longitude, station_b.latitude,
                                               station_b.longitude)
                distances[i].append(dist)
            
        return distances
```

```python
class ServerPlacer(object):
    def __init__(self, base_stations: List[BaseStation], distances: List[List[float]]):
        self.base_stations = base_stations.copy()
        self.edge_servers = None
        self.distances = distances

       

    def _distance_edge_server_base_station(self, edge_server: EdgeServer, base_station: BaseStation) -> float:
        
        if edge_server.base_station_id:
            return self.distances[edge_server.base_station_id][base_station.id]
        return DataUtils.calc_distance(edge_server.latitude, edge_server.longitude, base_station.latitude,
                                       base_station.longitude)

    def compute_objectives(self):
        objectives = {
            'latency': self.objective_latency(), 
            'workload': self.objective_workload()
        }
        return objectives

    def objective_latency(self):
        
        assert self.edge_servers
        total_delay = 0
        base_station_num = 0
        for es in self.edge_servers:
            for bs in es.assigned_base_stations:
                delay = self._distance_edge_server_base_station(es, bs)
                
                total_delay += delay
                base_station_num += 1
        return total_delay / base_station_num

    def objective_workload(self):
       
        assert self.edge_servers
        workloads = [e.workload for e in self.edge_servers]
        
        res = np.std(workloads)
        return res
    
    
    def print_edge_servers_info(self):
        if not hasattr(self, 'edge_servers'):
            print("Edge servers have not been placed yet. Please run place_server first.")
            return
        for i, edge_server in enumerate(self.edge_servers):
            print(f"Edge Server {i}:")
            print(f"  Location: Latitude {edge_server.latitude}, Longitude {edge_server.longitude}")
            print(f"  Assigned Base Stations: {len(edge_server.assigned_base_stations)}")
            print(f"  Total Workload: {edge_server.workload}")
            # If you also want to print the IDs of the assigned base stations:
            assigned_ids = [bs.id for bs in edge_server.assigned_base_stations]
            print(f"  Assigned Base Station IDs: {assigned_ids}\n")
```

```python
class KMeansServerPlacer(ServerPlacer):
    """
    K-means approach
    """
    name = 'KMeans'
    def place_server(self, base_station_num, edge_server_num):
  
        # init data as ndarray
        base_stations = self.base_stations[:base_station_num]
        coordinates = list(map(lambda x: (x.latitude, x.longitude), base_stations))
        data = np.array(coordinates)
        k = edge_server_num

        # k-means
        kmeans = KMeans(n_clusters=k, random_state=0, max_iter=100).fit(data)
        centroid = kmeans.cluster_centers_
        label = kmeans.labels_


        # # k-means
        # kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0, max_iter=100).fit(data)
        # centroid = kmeans.cluster_centers_
        # label = kmeans.labels_

        # process result
        edge_servers = [EdgeServer(i, row[0], row[1]) for i, row in enumerate(centroid)]
        for bs, es in enumerate(label):
            edge_servers[es].assigned_base_stations.append(base_stations[bs])
            edge_servers[es].workload += base_stations[bs].workload

        self.edge_servers = list(filter(lambda x: x.workload != 0, edge_servers))



    def refine_edge_servers_recursively(self, edge_server, min_assigned_stations):
        if len(edge_server.assigned_base_stations) <= min_assigned_stations:
            return [edge_server]

        coordinates = np.array([(bs.latitude, bs.longitude) for bs in edge_server.assigned_base_stations])
        if len(coordinates) <= 1:  # Đảm bảo có đủ điểm để chạy K-means
            return [edge_server]

        kmeans = KMeans(n_clusters=2, random_state=0).fit(coordinates)
        centroids = kmeans.cluster_centers_
        new_edge_servers = []

        for i, centroid in enumerate(centroids):
            new_id = f"{edge_server.id}.refined.{i+1}"
            assigned_stations = [edge_server.assigned_base_stations[j] for j in range(len(coordinates)) if kmeans.labels_[j] == i]
            workload = sum(bs.workload for bs in assigned_stations)
            new_server = EdgeServer(new_id, centroid[0], centroid[1])
            new_server.assigned_base_stations = assigned_stations
            new_server.workload = workload
            # Gọi đệ quy cho mỗi máy chủ mới
            new_edge_servers.extend(self.refine_edge_servers_recursively(new_server, min_assigned_stations))

        return new_edge_servers

    def refine_edge_servers(self, min_assigned_stations):
        new_edge_servers = []
        for edge_server in self.edge_servers:
            new_edge_servers += self.refine_edge_servers_recursively(edge_server, min_assigned_stations)
        self.edge_servers = new_edge_servers


    def print_edge_servers_info(self):
        if not hasattr(self, 'edge_servers'):
            print("Edge servers have not been placed yet. Please run place_server first.")
            return
        for i, edge_server in enumerate(self.edge_servers):
            print(f"Edge Server {i}:")
            print(f"  Location: Latitude {edge_server.latitude}, Longitude {edge_server.longitude}")
            print(f"  Assigned Base Stations: {len(edge_server.assigned_base_stations)}")
            print(f"  Total Workload: {edge_server.workload}")
            assigned_ids = [bs.id for bs in edge_server.assigned_base_stations]
            print(f"  Assigned Base Station IDs: {assigned_ids}\n")
```

```python
class TopKServerPlacer(ServerPlacer):
    """
    Top-K approach
    """

    def place_server(self, base_station_num, edge_server_num):
        
        base_stations = self.base_stations[:base_station_num]
        sorted_base_stations = sorted(base_stations, key=lambda x: x.workload, reverse=True)
        edge_servers = [EdgeServer(i, item.latitude, item.longitude, item.id) for i, item in
                        enumerate(sorted_base_stations[:edge_server_num])]
        for i, base_station in enumerate(sorted_base_stations):
            closest_edge_server = None
            min_distance = 1e10
            for j, edge_server in enumerate(edge_servers):
                tmp = self._distance_edge_server_base_station(edge_server, base_station)
                if tmp < min_distance:
                    min_distance = tmp
                    closest_edge_server = edge_server
            closest_edge_server.assigned_base_stations.append(base_station)
            closest_edge_server.workload += base_station.workload
        self.edge_servers = edge_servers
      
    def print_edge_servers_info(self):
        if not hasattr(self, 'edge_servers'):
            print("Edge servers have not been placed yet. Please run place_server first.")
            return
        for i, edge_server in enumerate(self.edge_servers):
            print(f"Edge Server {i}:")
            print(f"  Location: Latitude {edge_server.latitude}, Longitude {edge_server.longitude}")
            print(f"  Assigned Base Stations: {len(edge_server.assigned_base_stations)}")
            print(f"  Total Workload: {edge_server.workload}")
            # If you also want to print the IDs of the assigned base stations:
            assigned_ids = [bs.id for bs in edge_server.assigned_base_stations]
            print(f"  Assigned Base Station IDs: {assigned_ids}\n")
```
